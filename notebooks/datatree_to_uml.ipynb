```python
import xarray as xr
import datatree
from pathlib import Path
from xarray_eop.eop import datatree_to_uml
from xarray_eop.eop import open_eop_datatree
```

```python
path=Path("/mount/internal/work-st/projects/cs-412/2078-dpr/Samples/Products/Zarr/NewL0Products_2024-05-29")

for p in path.glob("S03*"):
    print(p)
    dt = datatree.open_datatree(p,chunks={},engine="zarr")
    pattern=p.name[:10]
    uml_filename = "_".join([pattern,"product","structure"])+".puml"
    uml = datatree_to_uml(dt,name=p.name[:9])
    with open(uml_filename,"w") as f:
        f.write(uml)
```

```python
import s3fs
import json
# Assuming credentials are in ~/.eopf/secrets.json
try:
    SECRET_PATH = Path.home() / Path(".eopf/secrets.json")
    with open(SECRET_PATH) as f:
        secrets=json.load(f)

# Or use ENV variables
except:
    import os
    secrets = {"s2input" : {
        "key": os.getenv("AWS_ACCESS_KEY_ID"),
        "secret": os.getenv("AWS_SECRET_ACCES_KEY"),
        "endpoint_url": os.getenv("AWS_ENDPOINT_URL"),
        "region_name": os.getenv("AWS_DEFAULT_REGION")
        }
    }

secrets["s2input"].pop("region_name",None)

SAMPLE_PATH = "s3://s2-input/Products/"
s3 = s3fs.S3FileSystem(
    key=secrets["s2input"]["key"],
    secret=secrets["s2input"]["secret"],
    endpoint_url=secrets["s2input"]["endpoint_url"]
)
s3_path = s3.glob(SAMPLE_PATH+"S02*.zarr")
store=[f"s3://{p}" for p in s3_path]
store
```

```python
for p in store[1:]:
    print(p)
    dt = open_eop_datatree(p,backend_kwargs={"storage_options": {"s3":secrets["s2input"]}})
    pattern=p.split("/")[-1][:9]
    uml_filename = "_".join([pattern,"product","structure"])+".puml"
    uml = datatree_to_uml(dt,name=pattern)
    with open(uml_filename,"w") as f:
        f.write(uml)
```
